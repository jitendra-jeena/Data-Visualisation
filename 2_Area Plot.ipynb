# Visualizing Data using Matplotlib

```python
#import necessary modules
import matplotlib.pyplot as plt

# use the inline backend to generate the plots within the browser
%matplotlib inline 
```

# Area Plots

An area chart represents the change in one or more quantities over time. It is similar to a line graph. In both area charts and line graphs, data points are plotted and then connected by line segments to show the value of a quantity at several different times. Area charts are different from line graphs, however, because the area between the x axis and the line is filled in with color or shading.

# Stacked Area Charts

Sometimes, you may need to represent how several quantities combine to make a whole. A stacked area chart is great at this! A stacked area chart shows how much each part contributes to the whole amount. 

Area plots are stacked by default. And to produce a stacked area plot, each column must be either all positive or all negative values (any NaN values will defaulted to 0). To produce an unstacked plot, pass `stacked=False`. 

```python
# Lets visualise the contribution of top 5 countries 
```

```python
years = list(map(str,range(1980,2014)))
```

```python
#Sorting dataset on the basis of total immigration
df_canada.sort_values(by = "Total",ascending = False,inplace = True)
```

```python
df_top5 = df_canada.head()
#Extract top 5 countries
#view top 2 rows
df_top5.head(2)
```

```python
#Extrat the relevant columns i.e from year 1980 to 2013
# transpose it as pandas use index as x_values
df_top5 = df_top5[years].transpose()

#view top 2 rows
df_top5.head(2)
```

```python
#Ploat the area plot
#by passing parameter kind  = "area" 
df_top5.plot(kind = "area",
            stacked = False,
            figsize = (20,10))

plt.title('Immigration Trend of 5 Countries with Most Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

The unstacked plot has a default transparency (alpha value) at 0.5. We can modify this value by passing in the `alpha` parameter.

```python
df_top5.index = list(map(int,df_top5.index))
df_top5.plot(kind='area', 
             alpha=0.8, # 0-1, default value a= 0.5
             stacked=True,
             figsize=(20, 10),
            )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
